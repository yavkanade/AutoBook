#MAIN RECURSIVE LOOP

````python

def book_chapter_output_system(context_string="No additional content provided",
                                     previous_chapter_summaries=None, previous_state=None,
                                     prompt_example=None, task="None", user_comment=None,
                                     chapter=1, chapter_breakdown="None", book="",
                                     total_chapters=15,
                                     cold=None, plot_example=None):
    print("Start of function")
    print(f'\n\n\n\n')

    print("+"*100)
    print("Book")
    print("+"*100)
    print(book)
    print("+"*100)
    print(f'\n\n\n\n')

    try:
        print("Inside try block")
        n = 1
        min = 990
        max = 1200
        best_prompt = None
        usr="no"
        if chapter_breakdown == "None":
            while usr=="no":
                print("Chapter breakdown is None. Generating plot.")
                n = 3
                plot_instructions = get_plot_instructions(task)
                if plot_example is not None:
                    extra_instructions=f"""Example of output required from you[{plot_example}]

                """
                    plot_instructions=plot_instructions+extra_instructions

                print(f"Plot instructions: {plot_instructions}")
                temperature=random.uniform(0.01, 0.2)
                plot = get_completion(plot_instructions, model="gpt-3.5-turbo", temperature=temperature)
                chapter_breakdown = plot
                print_chapter_breakdown(chapter_breakdown)
                print('temp:',temperature)
                usr=input("Do you want to continue? (yes/no)")


        print("Setting temperatures")
        if cold is None:
            ag1_temperatures = np.sort(np.random.normal(0.1, 0.04, n))
        else:
            ag1_temperatures = (np.random.normal(cold, 0.0, n))
        ag2_temperature = 0.1

        print("Getting current chapter")
        current_chapter=get_completion(f' You are agent responsible for keeping track of the current state of the story. Current chapter is Chapter {chapter}, you must only produce output with all of the information about this chapter. Chapter {chapter}: [{chapter_breakdown}]')
        print(f"Current chapter: {current_chapter}")
        print("-"*100)

        print("Generating AG1 instruction")
        ag1_instruction = generate_ag1_instruction(task, current_chapter, previous_state, prompt_example, user_comment, min, max, plot=chapter_breakdown)

        ag1_responses, ag2_responses = [], []
        print("Entering AG1 and AG2 generation loop")


        for i, temp in enumerate(ag1_temperatures, start=1):
            print(f"Iteration {i}")
            ag1_output = get_completion(ag1_instruction, temperature=temp)
            ag1_responses.append(ag1_output)
            ag2_instruction = generate_ag2_instruction(ag1_output, previous_chapter_summaries, context_string,chapter, total_chapters)
            ag2_output = get_completion(ag2_instruction, model="gpt-3.5-turbo", temperature=ag2_temperature)
            ag2_responses.append(ag2_output)
            print(100 * '=>')
            print(f"Agent 2 results {i}  temp={temp}")
            print(100 * '=')
            print(f"Option {i}: {ag2_output}")

        #selected_option=int(random.uniform(0,1)+1)
        #selected_option = int(input("Select an option (1-n) or input 0 to exit: "))
        #user_comment = input("Please provide a comment about the outcome: ")
        if chapter == 1:
            selected_option = int(input("Select an option (1-n) or input 0 to exit: "))
        else:
            selected_option=1
        chapter+=1

        if selected_option != 0:

            if chapter == 1:
                return book_chapter_output_system(context_string=ag2_responses[selected_option - 1], task=task, prompt_example=best_prompt,
                                                  chapter=chapter, chapter_breakdown=chapter_breakdown, book=book,total_chapters=total_chapters)

            if chapter == total_chapters+1:
                book= book + ag2_responses[selected_option - 1] + "\n"+ "\n"+ "\n"+ "\n"
                return book, chapter_breakdown
            best_prompt = ag1_responses[selected_option - 1]
            cold=ag1_temperatures[selected_option - 1]
            book= book + ag2_responses[selected_option - 1] + "\n"+ "\n"+ "\n"+ "\n"
            return book_chapter_output_system(context_string, previous_chapter_summaries=current_chapter, prompt_example=best_prompt, task=task, user_comment=user_comment, chapter=chapter, chapter_breakdown=chapter_breakdown, book=book, total_chapters=total_chapters, cold=cold)
        else:
            book= book + ag2_responses[selected_option - 1]
            return book, chapter_breakdown
        pass
    except Exception as e:  # Replace Exception with the specific error class, if known
        print(f"An error occurred: {e}")
        print("Retrying...")

        if chapter == 1:
            return book_chapter_output_system(context_string, task=task, prompt_example=best_prompt,
                                                  chapter=chapter, chapter_breakdown=chapter_breakdown, book=book,total_chapters=total_chapters)

        else:
            return book_chapter_output_system(context_string, previous_chapter_summaries=current_chapter, prompt_example=best_prompt, task=task, user_comment=user_comment, chapter=chapter, chapter_breakdown=chapter_breakdown, book=book,total_chapters=total_chapters,cold=cold)
##############
````
